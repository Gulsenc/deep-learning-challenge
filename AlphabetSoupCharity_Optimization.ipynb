```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])

application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()

unique_values
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
replace_types = application_type_counts[application_type_counts < 1000].index
application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(replace_types, 'Other')
print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 1000
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < cutoff_value].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()

print(classification_counts)

cutoff_value = 1000
classifications_to_replace = classification_counts[classification_counts < cutoff_value].index

for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")


print(application_df['CLASSIFICATION'].value_counts())
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = application_df['CLASSIFICATION'].value_counts()

classification_counts_gt_1 = classification_counts[classification_counts > 1]

print(classification_counts_gt_1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1000  # You can adjust this value based on your desired threshold
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() < cutoff_value].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)

application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays

X = application_df.drop(columns=['IS_SUCCESSFUL'])
y = application_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the optimized model
nn_optimized = tf.keras.models.Sequential()

# First hidden layer with increased neurons
nn_optimized.add(tf.keras.layers.Dense(units=200, activation='relu', input_dim=X_train.shape[1]))

# Second hidden layer with increased neurons
nn_optimized.add(tf.keras.layers.Dense(units=100, activation='relu'))

# Third hidden layer with increased neurons
nn_optimized.add(tf.keras.layers.Dense(units=50, activation='relu'))

# Fourth hidden layer added
nn_optimized.add(tf.keras.layers.Dense(units=25, activation='relu'))

# Output layer
nn_optimized.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the optimized model
nn_optimized.summary()

# Compile the optimized model
nn_optimized.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the optimized model with increased epochs
fit_model_optimized = nn_optimized.fit(X_train_scaled, y_train, epochs=100, validation_split=0.2, callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)])

# Evaluate the optimized model using the test data
model_loss_optimized, model_accuracy_optimized = nn_optimized.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Optimized Loss: {model_loss_optimized}, Optimized Accuracy: {model_accuracy_optimized}")
```

```python
# Define the further optimized model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn_further_optimized = tf.keras.models.Sequential()

# First hidden layer with increased neurons and batch normalization
nn_further_optimized.add(tf.keras.layers.Dense(units=200, activation='relu', input_dim=X_train.shape[1]))
nn_further_optimized.add(tf.keras.layers.BatchNormalization())
nn_further_optimized.add(tf.keras.layers.Dropout(0.2))

# Second hidden layer with increased neurons, batch normalization, and dropout
nn_further_optimized.add(tf.keras.layers.Dense(units=100, activation='relu'))
nn_further_optimized.add(tf.keras.layers.BatchNormalization())
nn_further_optimized.add(tf.keras.layers.Dropout(0.2))

# Third hidden layer with increased neurons, batch normalization, and dropout
nn_further_optimized.add(tf.keras.layers.Dense(units=50, activation='relu'))
nn_further_optimized.add(tf.keras.layers.BatchNormalization())
nn_further_optimized.add(tf.keras.layers.Dropout(0.2))

# Fourth hidden layer
nn_further_optimized.add(tf.keras.layers.Dense(units=25, activation='relu'))
nn_further_optimized.add(tf.keras.layers.BatchNormalization())
nn_further_optimized.add(tf.keras.layers.Dropout(0.2))

# Output layer
nn_further_optimized.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the further optimized model with a different optimizer and learning rate
nn_further_optimized.compile(loss="binary_crossentropy", optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=["accuracy"])

# Train the further optimized model with increased epochs and early stopping
fit_model_further_optimized = nn_further_optimized.fit(X_train_scaled, y_train, epochs=100, validation_split=0.2, callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)])

# Evaluate the further optimized model using the test data
model_loss_further_optimized, model_accuracy_further_optimized = nn_further_optimized.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Further Optimized Loss: {model_loss_further_optimized}, Further Optimized Accuracy: {model_accuracy_further_optimized}")
```

```python
from tensorflow.keras.layers import LeakyReLU

# Define the more complex and optimized model
nn_more_optimized = tf.keras.models.Sequential()

# First hidden layer with LeakyReLU activation
nn_more_optimized.add(tf.keras.layers.Dense(units=256, input_dim=X_train.shape[1], kernel_initializer='he_normal'))
nn_more_optimized.add(LeakyReLU(alpha=0.1))
nn_more_optimized.add(tf.keras.layers.BatchNormalization())
nn_more_optimized.add(tf.keras.layers.Dropout(0.3))

# Second hidden layer with LeakyReLU activation
nn_more_optimized.add(tf.keras.layers.Dense(units=128, kernel_initializer='he_normal'))
nn_more_optimized.add(LeakyReLU(alpha=0.1))
nn_more_optimized.add(tf.keras.layers.BatchNormalization())
nn_more_optimized.add(tf.keras.layers.Dropout(0.3))

# Third hidden layer with LeakyReLU activation
nn_more_optimized.add(tf.keras.layers.Dense(units=64, kernel_initializer='he_normal'))
nn_more_optimized.add(LeakyReLU(alpha=0.1))
nn_more_optimized.add(tf.keras.layers.BatchNormalization())
nn_more_optimized.add(tf.keras.layers.Dropout(0.3))

# Fourth hidden layer
nn_more_optimized.add(tf.keras.layers.Dense(units=32, kernel_initializer='he_normal'))
nn_more_optimized.add(LeakyReLU(alpha=0.1))
nn_more_optimized.add(tf.keras.layers.BatchNormalization())
nn_more_optimized.add(tf.keras.layers.Dropout(0.3))

# Output layer
nn_more_optimized.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the model with a learning rate scheduler
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=10000,
    decay_rate=0.9)
optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

nn_more_optimized.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])

# Train the optimized model
fit_model_more_optimized = nn_more_optimized.fit(X_train_scaled, y_train, epochs=150, validation_split=0.2, callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)])

# Evaluate the optimized model using the test data
model_loss_more_optimized, model_accuracy_more_optimized = nn_more_optimized.evaluate(X_test_scaled, y_test, verbose=2)
print(f"More Optimized Loss: {model_loss_more_optimized}, More Optimized Accuracy: {model_accuracy_more_optimized}")
```

```python
# Export our model to HDF5 file
nn_more_optimized.save("AlphabetSoupCharity_Optimization.h5")
```
